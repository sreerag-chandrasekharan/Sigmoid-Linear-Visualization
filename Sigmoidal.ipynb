# Sigmodial Function

To impleiment and understat the working of sigmoidal fi=unction

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data

```python
# Generate an array of evenly spaced values between -10 and 10
x = np.arange(-10,11)
y = np.arange(-15, 26, 2)
np.shape(x)
```

```python
w = 2
b = 5
```

### linear function

```python
f_x = w * x + b
print(f_x)
np.shape(f_x)
```

```python
# Plotting
plt.figure(figsize=(4, 3))
plt.plot(x, f_x, label=f' f_w' )
plt.scatter(x,y,color='red', marker='o')
plt.title('Plot of Linear Function y = 2x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)  # x-axis
plt.axvline(0, color='black',linewidth=1)  # y-axis
plt.legend()
plt.show()
```

### Sigmoidal function

```python
Z = w * x + b
```

```python
def sigmoid(z):
    
    g = 1/(1+np.exp(-z))
   
    return g
```

```python
f_x = sigmoid(Z)
```

```python
print(f_x)
np.shape(f_x)
```

```python
# Generate subplots for f(x) by Z and f(x) by x
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3))

# Plot f(x) by Z
ax1.plot(Z, f_x, label='Sigmoid(Z)', color='green', marker='o')
ax1.set_title('Sigmoid Function applied to Z = 2x + 5')
ax1.set_xlabel('Z')
ax1.set_ylabel('f(Z) = sigmoid(Z)')
ax1.grid(True)
ax1.axhline(0, color='black', linewidth=1)  # x-axis
ax1.axvline(0, color='black', linewidth=1)  # y-axis
ax1.legend()

# Plot f(x) by x
ax2.plot(x, f_x, label='Sigmoid(Z) by x', color='blue', marker='o')
ax2.set_title('Sigmoid Function by x')
ax2.set_xlabel('x')
ax2.set_ylabel('f(x) = sigmoid(Z)')
ax2.grid(True)
ax2.axhline(0, color='black', linewidth=1)  # x-axis
ax2.axvline(0, color='black', linewidth=1)  # y-axis
ax2.legend()

# Show plots
plt.tight_layout()
plt.show()
```

The decision bountry is at z = 0 , ie x = -2.5.
